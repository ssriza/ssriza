```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Create sample data: x is an array of 100 values between 0 and 10
x = np.linspace(0, 10, 100) # returns an array of 100 evenly spacedalues between 0 and 10 (inclusive)
# y is the sine of x
y = np.sin(x) #returns an array where each element is the sine of thecorresponding element in x
y2 = np.cos(x)
# Create the line plot
plt.plot(x, y, label='Sine Wave', color='black') # This creates theline plot by connecting the points (x, y) with a line.
plt.plot(x, y2, label='coSine Wave', color='red')
# Add title to the plot
plt.title('Line Plot - Sine Wave')
# Label the x-axis
plt.xlabel('X-axis')
# Label the y-axis
plt.ylabel('Y-axis')
# Add a legend in the upper right corner
plt.legend()
# Display a grid to help with readability of the plot
plt.grid(True)
# Show the plot
plt.show()
```

```python
import seaborn as sns
import numpy as np

data = np.random.normal(size=1000)
sns.histplot(data, bins=25,kde=True,color='blue')
plt.title('Histogram with KDE')
plt.show()
```

```python
#box plot
data= np.random.normal(size=100)
plt.boxplot(data)
plt.title('box plot of data')
plt.ylabel('data')
plt.xlabel('x')
plt.show()
```

```python
import pandas as pd
# Create a dataset for comparison
data = pd.DataFrame({
 'Age': [23, 45, 34, 42, 25, 36, 52, 39, 40, 30],
 'Salary': [40000, 80000, 60000, 75000, 42000, 62000, 95000, 71000,
85000, 53000]
})
```

```python
sns.scatterplot(x=data['Age'], y=data['Salary'],color='green')
plt.title('Age vs Salary (Scatter plot)')
plt.show()

#histogram
sns.histplot(data['Age'], bins = 5, kde = True, color='purple')
plt.title('Histogram of Age')
plt.show()

# box plot salary
sns.boxplot(data['Salary'], color ='orange')
plt.title('boxplot of salary')
plt.show()
```

```python
#Creating all 3 in the same window for comparison
# Create a figure with 1 row and 3 columns
plt.figure(figsize=(15, 5)) # This creates a figure with a specific size, here 15 inches wide and 5 inches tall, to accommodate threeplots side by side.

#scatterplot
plt.subplot (1,3,1)
sns.scatterplot (x=data['Age'], y=data['Salary'], color ='purple')
plt.title('Age vs Salary (Scatter Plot)')
plt.xlabel('Age')
plt.ylabel('Salary')
# Histogram (2nd subplot)
plt.subplot(1, 3, 2) # (rows, columns, position)
sns.histplot(data['Age'], bins=5, kde=True, color='green')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
# Box Plot (3rd subplot)
plt.subplot(1, 3, 3) # (rows, columns, position)
sns.boxplot(data['Salary'], color='orange')
plt.title('Box Plot of Salary')
plt.xlabel('Salary')

plt.tight_layout()
plt.show()
```
