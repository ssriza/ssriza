```python
import numpy as np
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
X = np.array([3],[4],[6],[7],[8],[10],[12]])
y = np.array([50, 60, 80, 95, 110, 130 , 150])
model = LinearRegression()
model.fit(X, y)
```

```python
data = {
    'Size': [800, 1000, 1200, 1500, 1800, 2000, 2200, 2500],
    'Price': [150000, 200000, 240000, 290000, 330000, 360000, 400000, 440000]
}
df= pd.DataFrame(data)
```

```python
np.random.seed(0)
house_sizes = np.random.randint(500, 2500, size = 100)
house_prices = 50 * house_sizes + np.random.normal(0,10000, size=100)
house_prices

```

```python
X = df[['Size']]
y = df['Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
LinearRegression()
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
# visualize the results
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color ='red', label='Model Fit')
plt.xlabel("Size (sq ft)")
plt.ylabel("Price(USD)")
plt.title("House Size vs. Price Prediction")
plt.legend()
plt.show()
```
